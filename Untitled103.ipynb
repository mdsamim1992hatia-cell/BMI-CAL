```python
df1=pd.DataFrame({
    'a':[1,2,np.nan],
    'b':[5,np.nan,np.nan],
    'c':[1,None,3]
})
df=pd.DataFrame(df1)
df
```

```python
df2=pd.DataFrame({
    'a':[1,2,3],
    'b':[5,6,7],
    'c':[1,2,3]
})
df=pd.DataFrame(df2)
df
```

```python
result=pd.concat([df,df2])
result
```

```python
result=pd.merge(df,df2,how='inner',on='a')
result
```

```python
result=pd.merge(df,df2,how='outer',on='a')
result
```

```python
result=pd.merge(df,df2,how='left',on='a')
result
```

```python
result=pd.merge(df,df2,how='right',on='a')
result
```

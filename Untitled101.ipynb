```python
df1=pd.DataFrame({
    'key':['daka','newdeli',np.nan],
    'age':[50,555,55],
    'emply':[1,np.nan,3]
})
df1
```

```python
df2=pd.DataFrame({
    'key':['daka','newdeli',np.nan],
    'age':[50,555,55],
    'emply':[1,np.nan,3]
})
df2
```

```python
result=pd.concat([df1,df2])
result
```

```python
result=pd.merge(df1,df2,how='inner',on='key')
result
```

```python
result=pd.merge(df1,df2,how='outer',on='key')
result
```

```python
result=pd.merge(df1,df2,how='left',on='key')
result
```

```python
result=pd.merge(df1,df2,how='right',on='key')
result
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({
    'store':['a','b','c','d'],
    'sales':[400,5555,555,444],
    'profit':[100,200,300,400]
})
df=pd.DataFrame(df)
df
```

```python
df.groupby('store')['sales'].agg(['sum','mean','max'  ])
```

```python
df=pd.DataFrame({
    'customer_id':[1,2,3,4],
    'sales':[100,200,30,545]
})
df=pd.DataFrame(df)
df
```

```python
df.groupby('customer_id')['sales'].sum().sort_values(ascending=False)
```

```python
df=pd.DataFrame({
    'deft':['a','b','c','d'],
    'salary':[5552,888,444,555]})
df.groupby('deft')['salary'].describe()
```

```python
df=pd.DataFrame({
    'day':['mon','wed','thu','fri'],
    'sales':[400,5500,555,8585]})
df.groupby('day')['sales'].sum()
```

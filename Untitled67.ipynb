```python
import pandas as pd
import numpy as np
```

```python
data={
    "a":[21,5,45,np.nan,66],
    "b":[89,78,88,66,np.nan],
    "c":[88,99,66,np.nan,np.nan],
    "d":[88,919,66,np.nan,np.nan],
    "e":[88,99,616,np.nan,np.nan]
}
df=pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
df.ffill=df.fillna(method="ffill")
df.ffill
```

```python
df.bfill=df.fillna(method="bfill")
df.bfill
```

```python
df.zero=df.fillna(0)
df.zero
```

```python
def totall_nan(df):
    return df.isnull().sum().sum()
totall_nan(df)
```

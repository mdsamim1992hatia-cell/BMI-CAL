```python
df1=pd.DataFrame({
    'KEY':['K0','K1','K2','K3'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
})
df1
```

```python
df2=pd.DataFrame({
    'KEY':['K0','K1','K2','K3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
df2
```

```python
result=pd.merge(df1,df2,how='inner',on='KEY')
result
```

```python
result=pd.merge(df1,df2,how='outer',on='KEY')
result
```

```python
result=pd.merge(df1,df2,how='right',on='KEY')
result
```

```python
result=pd.merge(df1,df2,how='right',on='KEY')
result
```

```python
result=pd.merge(df1,df2,how='right',on='KEY')
result
```

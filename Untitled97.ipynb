```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({
    'store':['dhaka','dhaka','khulna','khulna','rajsahi','rajsahi'],
    'custumer':['tom','jerry','tom','jerry','tom','jerry'],
    'sales':[500,555,600,655,700,755]
})
df=pd.DataFrame(df)
df
```

```python
df.groupby('store')[['sales']].mean()
```

```python
df.groupby('store')[['sales']].sum()
```

```python
by_store=df.groupby('store')
by_store.describe()
```

```python
by_store.sum()
```

```python
df.groupby('store').count()
```

```python
df.groupby('store').max()
```

```python
df.groupby('store').min()
```

```python
import pandas as pd
import numpy as np
```

```python
array_2d=np.array([[1,2,np.nan,45,3],[4,5,np.nan,6,9],[7,8,np.nan,9,5]])
print(array_2d )
```

```python
df=pd.DataFrame(array_2d)
print(df)
```

```python
df_bfill=df.fillna(method='bfill')
print(df_bfill)
```

```python
df_ffill=df.fillna(method='ffill')
print(df_ffill)
```

```python
df_zero=df.fillna(0)
print(df_zero)
```

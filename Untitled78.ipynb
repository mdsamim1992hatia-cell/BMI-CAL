```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({
    'key':['a','b','c','d','e'],
    'a1':range(5),
    'b1':range(5,10)
})
df1
```

```python
df2=pd.DataFrame({
    'key':['a','b','c'],
    'a2':range(3),
    'b2':range(3,6)
})
df2
```

```python
pd.merge(df1,df2,on='key')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='inner',on='key')[['key','a1','b1']]
```

```python
pd.merge(df1,df2,how='left',on='key')[['key','a2','b2']]
```

```python
pd.merge(df1,df2,how='right',on='key')[['key','a2','b2']]
```

```python
pd.merge(df1,df2,how='outer',on='key')[['key','a1','b1','a2','b2']]
```

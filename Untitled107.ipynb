```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({
    'store':['A','A','A','A','B','B','B','B','c','c','c','c'],
    'sales':[1000,788,88855,555,2222,5858,666,5554,858525,333,55,554]
})
df.groupby('store')['sales'].agg(['sum','min','max','mean'])
```

```python
df.groupby('store')['sales'].sum().sort_values(ascending=False)
```

```python
df=pd.DataFrame({
    'customer':['tom','jerry','moni','liliy'],
    'sales':[4556,8744,6555,2522]
})
df.groupby('customer')['sales'].sum()
```

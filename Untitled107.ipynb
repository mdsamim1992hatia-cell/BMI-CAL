```python
df1=pd.DataFrame({
    'a1':[10,20,30,40,50],
    'b1':[10,20,30,40,50],
    'c1':[10,20,30,40,50],
})
df1
```

```python
df2=pd.DataFrame({
    'a1':[10,20,30,40,50],
    'b2':[10,255 ,144,30,40],
    'c2':[10,20,30,40,50],
})
df2
```

```python
result=pd.concat([df1,df2],axis=1)
result
```

```python
result=pd.merge(df1,df2,how='inner',on='a1')
result
```

```python
result=pd.merge(df1,df2,how='outer',on='a1',)
result
```

```python
result=pd.merge(df1,df2,how='left',on='a1')
result
```

```python
result=pd.merge(df1,df2,how='right',on='a1')
result
```

```python
df=pd.DataFrame({
    'store':[1,2,1,2,1,2],
    'sales':[100,200,300,400,500,600]
})
df.groupby('store')['sales'].agg(['sum','max','min','mean'])
```

```python
df=pd.DataFrame({
    'customer':['c1','c2','c3','c4','c5'],
    'sales':[100,200,300,400,500]
})
df.groupby('customer')['sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('customer')['sales'].sum().sort_values(ascending=False).head(3)
```

```python
df=pd.DataFrame({
    'dept':['it','it','it','hr','hr','hr'],
    'selary':[100,200,300,400,500,600]
})
df.groupby('dept')['selary'].agg(['sum','max','min','mean'])
```

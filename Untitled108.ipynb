```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({
    'department':['hr','sales','it','hr','it'],
    'salery':[1000,88888,555,222,444,]
})
df.groupby('department')['salery'].describe()
```

```python
df=pd.DataFrame({
    'day':['sun','sun','mon','mon','tue','tue'],
    'sales':[1000,88888,555,222,4556,788]
})
df.groupby('day')['sales'].sum().sort_values(ascending=False)
```

```python
df=pd.DataFrame({
    'product':['a','b','a','b','a','b','a','a'],
    'sales':[1000,8988,252,6666,3333,444,8555,1111]
})
df.groupby('product')['sales'].agg(['sum','mean','count','max','min'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame({
    'store':['A','B','C','A','B','C','A','B','C'],
    'sales':[100,200,300,400,500,600,555,888,5555]
})
df.groupby('store').sum()
```

```python
df.groupby('store')['sales'].agg(['sum','mean','max','min'])
```

```python
df.groupby('store').agg({'sales':['sum','mean','max','min']})
```

```python
df=pd.DataFrame({
    'cusomer':['k','l','m','n','m','n','k'],
    'seales':[1010,20033,30550,40550,555,555,888]
})
df.groupby('cusomer')['seales'].agg(['sum','max','min','mean'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame({
    'store':['A2','A5','B7','B','C' ,'C'],
    'sales':[100,150,200,250,300,350]
})
df.groupby('store')['sales'].agg(['sum','mean','max','min'])
```

```python
df=pd.DataFrame({
    'customer':['A','A','B','B','C','C'],
    'sales':[100,150,200,250,300,350]
})
df
```

```python
df.groupby('customer')['sales'].sum().sort_values(ascending=False)
```

```python
df=pd.DataFrame({
    'dept':['A','A','B','B','C','C'],
    'salary':[5000,6000,7000,8000,9000,10000]
})
df
```

```python
df.groupby('dept')['salary'].mean()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.DataFrame({'key':['a','b','c','d','e'],
                  'a1':range(5),
                  'b1':range(5,10)})
df1
```

```python
df2=pd.DataFrame({'key':['a','b','c'],
      'a2':range(3),
      'b2':range(3,6)})
df2
```

```python
pd.merge(df1,df2,how='inner',on='key')[['key','a1','a2']]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.DataFrame({
    'key':['a','b','c','d','e'],
    'a1':range(5),
    'b1':range(5,10)
})
df1
```

```python
df2=pd.DataFrame({
    'key':['a','b','c'],
    'a2':range(3),
    'b2':range(3,6)
})
df2
```

```python
df1=pd.DataFrame({
    'key':['k1','k2','k3','k4'],
    'a1':['a11','a12','a13','a14'],
    'b1':['b11','b12','b13','b14']
})
df1


```

```python
df2=pd.DataFrame({
    'key':['k1','k2','k3','k5'],
    'a2':['a55','a56','a57','a58'],
    'b2':['b55','b56','b57','b58']
})
df2
```

```python
result=pd.merge(df1,df2,how='inner',on='key')[['key','a1','b1']]
result
```

```python
result=pd.merge(df1,df2,how='outer',on='key')
result
```

```python
result=pd.merge(df1,df2,how='left',on='key')
result
```

```python
result=pd.merge(df1,df2,how='right',on='key')
result
```

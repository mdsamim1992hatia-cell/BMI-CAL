```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({
    'city':['new york','dhaka','paris'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['new york','dhaka','paris'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
})
df2
```

```python
pd.merge(df1,df2,how='inner',on='city')
```

```python
df1=pd.DataFrame({
    'key':['k','l','m'],
    'c1':[2,0,1],
    'c2':[3,0,5]})
df1
```

```python
df2=pd.DataFrame({
    'key':['k','l','m'],
    'd1':[5,6,9],
    'd2':[1,3,5]
})
df2
```

```python
pd.merge(df1,df2,how='inner',on='key')[['key','c1','d1']]
```

```python
df3=pd.merge(df1,df2 ,how='left',on='key')[['key','c1','d1']]
df3
```

```python
df4=pd.merge(df1,df2,how='right',on='key')[['key','c1','c2']]
df4
```

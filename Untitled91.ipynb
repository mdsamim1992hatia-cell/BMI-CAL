```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'A':[1,2,3,4],
                 'b1':[5,6,7,8],
                 'c':[9,10,11,12],
})
df1
```

```python
df2=pd.DataFrame({'A':[1,2,3,4],
                  'b2':[5,6,7,8],
                  'c':[9,10,11,12],
})
df2
```

```python
pd.merge(df1,df2,how='inner',on='A')
```

```python
pd.merge(df1,df2,how='outer',on='A')
```

```python
pd.merge(df1,df2,how='left',on='A')
```

```python
pd.merge(df1,df2,how='right',on='A')
```

```python
pd.merge(df1,df2,how='outer',on=['A','c'])
```

```python
pd.merge(df1,df2,how='inner',on=['A','c'])
```
